```python
import pandas as pd

df = pd.read_csv('/content/Dataset .csv')
df.head()
```

```python
#Percentage of Table Booking
table_booking = df['Has Table booking'].value_counts(normalize=True) * 100
print(table_booking)
```

```python
#Percentage of Online Delivery
online_delivery = df['Has Online delivery'].value_counts(normalize=True) * 100
print(online_delivery)
```

```python
#Comparison of Ratings of Table Booking
booking_rating = df.groupby('Has Table booking')['Aggregate rating'].mean()
print(booking_rating)
```

```python
#Comparison of Ratings of Online Delivery
delivery_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean()
print(delivery_rating)
```

```python
#Comparison of Price Range and Online Delivery
price_delivery = pd.crosstab(df['Price range'], df['Has Online delivery'])
print(price_delivery)
```

Visualisation of data


```python
import matplotlib.pyplot as plt

ax = table_booking.plot(
    kind='bar',
    color=['darkblue', 'gold']
)

plt.title("Table Booking Availability (%)")
plt.xlabel("Table Booking")
plt.ylabel("Percentage")

# Add value labels
for i, v in enumerate(table_booking):
    ax.text(i, v + 1, f"{v:.1f}%", ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt

ax = online_delivery.plot(
    kind='bar',
    color=['darkblue', 'yellow']
)

plt.title("Online Delivery Availability (%)")
plt.xlabel("Online Delivery")
plt.ylabel("Percentage")

# Add value labels
for i, v in enumerate(online_delivery):
    ax.text(i, v + 1, f"{v:.1f}%", ha='center')

plt.show()
```

Comparison of Ratings

```python
import matplotlib.pyplot as plt

ax = booking_rating.plot(
    kind='bar',
    color=['tomato', 'seagreen']  # No = red, Yes = green
)

plt.title("Average Rating vs Table Booking")
plt.xlabel("Table Booking")
plt.ylabel("Average Rating")

# Add value labels on top of bars
for i, v in enumerate(booking_rating):
    ax.text(i, v + 0.05, f"{v:.2f}", ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt

ax = delivery_rating.plot(
    kind='bar',
    color=['steelblue', 'orange']
)

plt.title("Average Rating vs Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")

for i, v in enumerate(delivery_rating):
    ax.text(i, v + 0.05, f"{v:.2f}", ha='center')

plt.show()
```
