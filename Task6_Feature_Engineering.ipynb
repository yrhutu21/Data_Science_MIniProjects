```python
import pandas as pd

df = pd.read_csv('/content/Dataset .csv')
df.head()
```

```python
#New feature creation
df['Name Length'] = df['Restaurant Name'].apply(len) #Count no of characters
df[['Restaurant Name', 'Name Length']].head()
```

```python
#Creation of Address length
df['Address Length'] = df['Address'].apply(len)
df[['Address', 'Address Length']].head()
```

```python
#Table Booking
df['Has Table Booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df[['Has Table booking', 'Has Table Booking']].head()
```

```python
#Online Delivery
df['Has Online Delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
df[['Has Online delivery', 'Has Online Delivery']].head()
```

```python
#Number of Cuisines
df['Number of Cuisines'] = df['Cuisines'].apply(lambda x: len(str(x).split(',')))
df[['Cuisines', 'Number of Cuisines']].head()
```

```python
#Saving of updated dataset
df.to_csv("processed_dataset.csv", index=False)
```
