```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Dataset .csv')
df.head()
```

Most Common Price Ranges

```python
price_counts = df['Price range'].value_counts()
print(price_counts)
```

```python
#Visualisation of Price Ranges
ax = price_counts.plot(
    kind='bar',
    color='yellow',
    edgecolor='black'
)

plt.title("Most Common Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

for i, v in enumerate(price_counts):
    ax.text(i, v + 50, str(v), ha='center')

plt.show()
```

Average Rating per Price Range

```python
price_rating = df.groupby('Price range')['Aggregate rating'].mean()
print(price_rating)
```

```python
#Visualisation of Avg Rating per Price Range
ax = price_rating.plot(
    kind='bar',
    color=['lightblue', 'orange', 'green', 'purple'],
    edgecolor='black'
)

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

for i, v in enumerate(price_rating):
    ax.text(i, v + 0.05, f"{v:.2f}", ha='center')

plt.show()
```

Comparison of Color and Rating

```python
color_rating = df.groupby('Rating color')['Aggregate rating'].mean()
print(color_rating)
```

```python
#Visualtion of Color vs Rating
ax = color_rating.plot(
    kind='bar',
    color='skyblue',
    edgecolor='black'
)

plt.title("Rating Color vs Average Rating")
plt.xlabel("Rating Color")
plt.ylabel("Average Rating")

plt.show()
```
