```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
df = pd.read_csv('/content/Dataset .csv')
df.head()
```

```python
df.describe()
```

Mean , Median, Standard Derivation

```python
print("Mean :\n", df.mean(numeric_only = True)) #calculate mean
print("Median :\n", df.median(numeric_only = True)) #calculate median
print("Standard Deviation :\n", df.std(numeric_only = True)) #calculate standard deviation
```

Categorical Analysis

```python
df['City'].value_counts()
```

```python
df['Country Code'].value_counts()
```

```python
df['Cuisines'].value_counts().head(10)
```

```python
#Top Cities

top_cities = df['City'].value_counts().head(10)
top_cities
```

```python
#Top Cuisiness

top_cuisines = df['Cuisines'].value_counts().head(10)
top_cuisines
```

Visualisation of Top Cities and Top Cuisines Graph

```python
#Top Cities Graph

top_cities.plot(kind='bar', color='blue')
plt.title('Top Cities which are having most restaurents')
plt.xlabel('City')
plt.ylabel('Count')
#plt.xticks(rotation=45) #Rotate x-axis labels
plt.show()
```

```python
#For better visualisation of top cities

top_cities.plot(kind='bar')
plt.yscale('log')  # log scale
plt.title("Top Cities with Most Restaurants (Log Scale)")
plt.xlabel("City")
plt.ylabel("Log Count")
#plt.xticks(rotation=45)
plt.show()
```

```python
#Top Cuisines Graph

top_cuisines.plot(kind='bar', color='green')
plt.title('Top Cuisines which are having most restaurents')
plt.xlabel('Cuisines')
plt.ylabel('Count')
#plt.xticks(rotation=45) #Rotate x-axis labels
plt.show()
```
