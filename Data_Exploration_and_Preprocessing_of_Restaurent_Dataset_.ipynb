```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Dataset .csv") #Load dataset
```

```python
print(df.head()) #print first 5 rows
```

Dataset into Datatype Conversion

Check Missing Values

```python
print(df.isnull().sum())
```

Handle Missing Values

```python
df['Cuisines'].fillna('Unknown', inplace=True)
df.dropna(inplace=True)
```

Data Type Conversion

```python
print(df.dtypes)
```

```python
df['Aggregate rating'] = df['Aggregate rating'].astype(float)
```

Target Variable analyzing

```python
print(df['Aggregate rating'].describe())
```

```python
print(df['Aggregate rating'].value_counts())
```

Visualisation

```python
import matplotlib.pyplot as plt

df['Aggregate rating'].hist()
plt.title("Distribution of Ratings")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```
