```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Dataset .csv')
```

Rating Distribution

```python
plt.hist(df['Aggregate rating'], bins=20, edgecolor='black')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

Ratings by City

```python
city_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

ax = city_rating.plot(kind='bar', color='yellow', edgecolor='black')

plt.title("Top Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")

#plt.xticks(rotation=45)
plt.show()
```

Ratings by Cuisine

```python
df['Cuisines'] = df['Cuisines'].fillna('Unknown')
df_cuisine = df.assign(Cuisines=df['Cuisines'].str.split(',')).explode('Cuisines')
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].str.strip()

cuisine_rating = df_cuisine.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

ax = cuisine_rating.plot(kind='bar', color='purple', edgecolor='black')

plt.title("Top Cuisines by Rating")
plt.xlabel("Cuisine")
plt.ylabel("Average Rating")

#plt.xticks(rotation=45)
plt.show()
```

Votes vs Rating

```python
plt.scatter(df['Votes'], df['Aggregate rating'])
plt.title("Votes vs Rating")
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.show()
```

Price Range vs Rating

```python
price_rating = df.groupby('Price range')['Aggregate rating'].mean()

ax = price_rating.plot(kind='bar', color='purple', edgecolor='black')

plt.title("Price Range vs Rating")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

plt.show()
```
